```python

import pandas as pd
import re
import nltk
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import gensim
from gensim.models import Word2Vec
from sklearn.cluster import KMeans
```

#### Stopword removal, converting uppercase into lower case, and lemmatization

```python
stopwords = nltk.corpus.stopwords.words('english')
lemmatizer = WordNetLemmatizer()
```

```python
doc = []
for i in range(0, len(document)):
    d = re.sub('[^a-zA-Z]', ' ', document[i])
    d = d.lower()
    d = d.split()
    d = [lemmatizer.lemmatize(word) for word in d if not word in set(stopwords)]
    d = ' '.join(d)
    doc.append(d)
```

```python
doc[0]
```

#### Text Vectorization - TfIdf vectorization

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(doc)
```

```python
X.shape
```

```python
tf = pd.DataFrame(X.todense())
tf.columns = vectorizer.get_feature_names()
tf.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
Sum_of_squared_distances = []
K = range(2,10)
for k in K:
   km = KMeans(n_clusters=k, max_iter=200, n_init=10)
   km = km.fit(X)
   Sum_of_squared_distances.append(km.inertia_)
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
true_k = 5
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print ("Cluster %d:" % i)
    for ind in order_centroids[i, :20]:
        print(' %s' % terms[ind])
```

```python
lab = model.labels_
lab
```

```python
df = pd.DataFrame(list(zip(document, lab)), columns = ['Sentence', 'Cluster'])
df
```

```python
from wordcloud import WordCloud
result={'cluster':lab,'doc':document}
result=pd.DataFrame(result)
for k in range(0,true_k):
   s=result[result.cluster==k]
   text=s['doc'].str.cat(sep=' ')
   text=text.lower()
   text=' '.join([word for word in text.split()])
   wordcloud = WordCloud(max_font_size=50, max_words=100).generate(text)
   print('Cluster: {}'.format(k))
   plt.figure()
   plt.imshow(wordcloud, interpolation="bilinear")
   plt.axis("off")
   plt.show()

```

```python
import numpy as np
```

```python
d = []
for i in range(len(document)):
    doc = gensim.utils.simple_preprocess(document[i])
    doc = ' '.join(doc)
    d.append(doc)
```

```python
d
```

```python
model=Word2Vec(d, vector_size=50, min_count=2, sg=0)
```

```python
def vectorizer(s, m):
    features=[]
    for word in s:
        zero_vector = np.zeros(m.vector_size)
        vectors = []
        for w in word:
            if w in m.wv:
                vectors.append(m.wv[w])
        if vectors:
            vectors = np.asarray(vectors)
            avg_vec = vectors.mean(axis=0)
            features.append(avg_vec)
        else:
            features.append(zero_vector)
    return features
```

```python
# a=[]
# for i in d:
#     a.append(vectorizer(i,model))
# X = np.array(a)

v_d = vectorizer(d, model)
X= np.array(v_d)
```

```python
X
```

```python
true_k = 5
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit_predict(X)
```
